```python
def find_anomalies(data):
    anomalies = []
    random_data_std = statistics.stdev(data)
    random_data_mean = statistics.mean(data)
    anomaly_cut_off = random_data_std * 3
    lower_limit  = random_data_mean - anomaly_cut_off
    upper_limit = random_data_mean + anomaly_cut_off
    for outlier in data:
        if outlier > upper_limit or outlier < lower_limit:
            anomalies.append(outlier)
    return anomalies
```

```python
top_products = df.groupby('product')['sale_price'].max().sort_values(ascending=False).head(5)
```

```python
top_products.plot(kind='bar', color='skyblue')
plt.title("Top 5 Products with Highest Selling Price")
plt.xlabel("Product")
plt.ylabel("Selling Price")
plt.xticks(rotation=75,fontsize = 10)
plt.show()
```

```python
sale_per_cat = df.groupby('category')['sale_price'].sum().sort_values(ascending=False).head(5)
```

```python
plt.figure(figsize=(10, 6))
sale_per_cat.plot(kind='barh', color='purple')
plt.title('Total Sales by Category', fontsize=14)
plt.xlabel('Sale Price', fontsize=12)
plt.ylabel('Category', fontsize=12)
plt.show()
```
